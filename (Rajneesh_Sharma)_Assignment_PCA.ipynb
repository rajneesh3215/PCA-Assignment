```python
# Normalize the data
array=wine1.values
array
```

```python
wine1_norm=scale(array)
wine1_norm
```

# PCA

```python
# applying pca fit transform to dataset
pca=PCA(n_components=13)
wine_pca=pca.fit_transform(wine1_norm)
wine_pca
```

```python
# PCA component matrix or covariance matrix
pca.components_
```

```python
# The amount of variance that each pCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color='red')
```

```python
# Final DataFrame
final_df = pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['pc1','pc2','pc2'])], axis = 1)
```

```python
final_df
```

```python
import seaborn as sns
fig=plt.figure(figsize=(18,9))
sns.scatterplot(data=final_df)
```

# 1.Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python

dendrogram=sch.dendrogram(sch.linkage(wine1_norm,'complete'))
```

```python
plt.figure(figsize=(18,9))
dendrogram=sch.dendrogram(sch.linkage(wine1_norm,'single'))
```

```python
plt.figure(figsize=(18,9))
dendrogram=sch.dendrogram(sch.linkage(wine1_norm,'average'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine1_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine3=wine.copy()
wine3['clustersid']=hclusters.labels_
wine3
```

# K Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
# within cluster sum of squares criterian
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(wine1_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
# Cluster algorithm using k=3
clusters=KMeans(3).fit(wine1_norm)
clusters
```

```python
clusters.labels_
```

```python
wine2=wine.copy()
wine2['clusters3id']=clusters.labels_
wine2
```

```python
wine2['clusters3id'].value_counts()
```
